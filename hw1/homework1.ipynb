```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('./GLB.Ts_dSST.csv')
df.head()
```

## Transform data

Find average for each year

```python
new_data = []
columns = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

for index, row in df.iterrows():
  sum = 0
  for month in columns:
    sum += row[month]
  new_data.append([row['Year'], round(sum / 12, 3)])

new_df = pd.DataFrame(new_data, columns = ['Year', 'Average'])
```

```python
new_df.head()
```

```python
# Main chart settings
chart = alt.Chart(new_df).mark_point(
        color = '#5D5956',
        fill = '#F26C08',
        size = 100,
        opacity = 1,
        strokeWidth = 1).encode(
    x = alt.X(field='Year',
              title = 'Year',
              type = 'quantitative',
              scale = alt.Scale(
                  zero = False,
            ),
            axis = alt.Axis(
             format = 'k',
             tickCount = 15,
             )),
    y = alt.Y(field = 'Average',
              title='Average temperature per year (°C)',
              type = 'quantitative',
              scale = alt.Scale(
                  zero = False
              ),
            axis = alt.Axis(
             tickCount = 10,

             )),
    tooltip = [
        alt.Tooltip(field = 'Year', type = 'nominal'),
        alt.Tooltip(field = 'Average', title='Temperature (°C)', type = 'nominal')
    ]
)

# Zero line
line = alt.Chart(pd.DataFrame({'y': [0]})).mark_rule(color='#989898').encode(y='y')

# 1 Title line
title1 = alt.Chart({'values':[{'x': 1882, 'y': 0.95}]}).mark_text(
    text='Annual Global Surface Temperature,',
    align='left',
    fontSize=15,
    fontWeight=700
).encode(x='x:Q', y='y:Q')

# 2 Title line
title2 = alt.Chart({'values':[{'x': 1882, 'y': 0.89}]}).mark_text(
    text='Relative to Late 19th Century Average',
    align='left',
    fontSize=15,
    fontWeight=700
).encode( x='x:Q', y='y:Q')

# 'Hotter' text
hotter_text = alt.Chart({'values':[{'x': 2019, 'y': 0.05}]}).mark_text(
    align='right',
    text='HOTTER THAN THE 1880-1899 AVERAGE',
    fontWeight=500
).encode(x='x:Q', y='y:Q')

# 'Colder' text
colder_text = alt.Chart({'values':[{'x': 2019, 'y': -0.05}]}).mark_text(
    align='right',
    text='COLDER',
    fontWeight=500
).encode(x='x:Q', y='y:Q')

# Come together... right now!
alt.layer(chart, line, colder_text, hotter_text, title1, title2).properties(
    width = 950, height = 500, background = '#F9F9F9', padding = 25
    ).configure_axis(domain=False)
```
